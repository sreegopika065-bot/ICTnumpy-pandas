**QUESTION 1**

Convert the mpg column into a NumPy array and calculate:


*   The mean, median, and standard deviation of mpg.
*   The number of cars with mpg greater than 25

```python
import pandas as pd
import numpy as np
df = pd.read_csv('sample.csv')
mpg_array = df['mpg'].to_numpy()
mean_mpg = np.mean(mpg_array)
median_mpg = np.median(mpg_array)
std_mpg = np.std(mpg_array)
count_above_25 = np.sum(mpg_array > 25)
print(mean_mpg, median_mpg, std_mpg, count_above_25)
```

**QUESTION 2**

Using NumPy, filter all cars with more than 6 cylinders.
Return the corresponding car_name as a list


```python
import pandas as pd
import numpy as np
df = pd.read_csv('sample.csv')
df.columns = df.columns.str.strip().str.replace(" ", "_")
cyl_array = df['cylinders'].to_numpy()
car_names = df['car_name'].to_numpy()
filtered_cars = car_names[cyl_array > 6]
filtered_list = filtered_cars.tolist()
print(filtered_list)
```

**QUESTION 3**

Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('sample.csv')
df.columns = df.columns.str.strip().str.replace(" ", "_")
hp_array = df['horsepower'].replace('?', np.nan).to_numpy()
hp_array = hp_array.astype(float)
mean_hp = np.nanmean(hp_array)
hp_array[np.isnan(hp_array)] = mean_hp
updated_hp = np.round(hp_array * 1.10, 2)
print(updated_hp)
```

**QUESTION 4**

Find the average displacement of cars with an origin of 2 (Europe) using NumPy
indexing.

```python
displacement = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()
europe_disp = displacement[origin == 2]
avg_disp = np.mean(europe_disp)
print(avg_disp)
```

**QUESTION 5**

Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

unique_cyl = np.unique(cylinders)

mean_mpg_list = []

for c in unique_cyl:
    mean_value = float(round(np.mean(mpg[cylinders == c]), 2))
    mean_mpg_list.append((int(c), mean_value))

print(mean_mpg_list)
```
